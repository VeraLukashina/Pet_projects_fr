# Analyse des résultats du diplôme national du brevet (DNB) dans les collèges français

Ce projet a pour objectif d’accompagner les familles d’Étaules (Charente-Maritime) dans leur choix d’établissement, en identifiant les collèges les plus performants à proximité, sur la base des résultats au Diplôme National du Brevet (DNB).

L’analyse porte sur les résultats du DNB entre 2006 et 2021. Les performances des élèves sont étudiées selon plusieurs axes : évolution annuelle des taux de réussite, disparités régionales, comparaison entre établissements publics et privés, et focus particulier sur les collèges accessibles depuis la commune d’Étaules (17).


**Description des données:**

- `Session` : année de la session d'examen.
- `Numero d'etablissement` : identifiant UAI unique de l'établissement.
- `Type d'etablissement` : type d'établissement (COLLEGE, LYCEE PROFESSIONNEL, etc.).
- `Patronyme` : nom de l'établissement scolaire.
- `Secteur d'enseignement` : secteur auquel appartient l'établissement (PUBLIC ou PRIVÉ).
- `Commune` : code géographique de la commune d'implantation de l'établissement.
- `Libellé commune` : nom de la commune d'implantation de l'établissement.
- `Code département` : code géographique du département.
- `Libellé département` : nom du département d'implantation.
- `Code académie` : code géographique de l'académie.
- `Libellé académie` : nom de l'académie de l'établissement scolaire.
- `Code région` : code géographique de la région.
- `Libellé région` : nom de la région d'implantation.
- `Inscrits` : nombre d'élèves inscrits à l'examen.
- `Presents` : nombre de candidats présents aux épreuves du DNB.
- `Admis` : nombre de candidats admis au DNB.
- `Admis sans mention` : nombre de candidats admis sans mention (note moyenne entre 10 et 12 sur 20).
- `Nombre_d_admis_Mention_AB` : nombre de candidats admis avec la mention assez bien.
- `Admis Mention bien` : nombre de candidats admis avec la mention bien (note moyenne entre 14 et 16).
- `Admis Mention très bien` : nombre de candidats admis avec la mention très bien (note moyenne supérieure à 16).
- `Taux de réussite` : pourcentage de réussite, calculé en rapportant le nombre d'admis au nombre de présents.

Fonction pour convertir les noms de colonnes en snake_case

```python
def to_snake_case(s):
    s = unicodedata.normalize("NFD", s).encode("ascii", "ignore").decode("utf-8")  # Supprimer les accents
    s = s.lower()  # Convertir en minuscules
    s = re.sub(r"[^\w\d]+", "_", s)  # Remplacer les espaces, apostrophes et caractères spéciaux par _
    s = re.sub(r"__+", "_", s)  # Supprimer les _ répétitifs
    s = s.strip("_")  # Supprimer les _ en début et fin de chaîne
    return s
```

## Analyse exploratoire et descriptive des taux de réussite à l’échelle nationale.

Le taux de réussite est calculé comme le rapport entre le nombre d’admis et le nombre d’inscrits à l’examen. Cette mesure permet d’évaluer la proportion d’élèves ayant obtenu leur diplôme parmi l’ensemble des candidats inscrits.

Définir une fonction d'agrégation des données :

```python
def aggregate_data(df, group_columns, columns_to_show):
    """
    Agrège les données du DataFrame en fonction d'une colonne spécifiée et calcule les taux pour chaque catégorie
    d'admission. La fonction retourne un DataFrame contenant uniquement les colonnes demandées.

    Parameters:
    - df (pandas.DataFrame): Le DataFrame contenant les données à analyser.
    - group_columns (list): Une liste des colonnes par lesquelles effectuer le regroupement.
    - columns_to_show (list): Une liste des colonnes à afficher dans le DataFrame résultant. Les colonnes doivent inclure
      les taux calculés.

    Returns:
    - pandas.DataFrame: Un DataFrame avec les colonnes spécifiées dans `columns_to_show` après l'agrégation et le calcul
      des taux d'admission.
    """
    aggregated = (
        df.groupby(group_columns)[
            ["admis", "admis_sans_mention", "nombre_d_admis_mention_ab", "admis_mention_bien", "admis_mention_tres_bien", 
             "presents", "inscrits"]
        ]
        .sum()
        .assign(
            taux_presents = lambda x: x["presents"] / x["inscrits"],
            taux_admis = lambda x: x["admis"] / x["inscrits"],
            taux_admis_sm = lambda x: x['admis_sans_mention'] / x['inscrits'],
            taux_admis_ab = lambda x: x['nombre_d_admis_mention_ab'] / x['inscrits'],
            taux_admis_b = lambda x: x['admis_mention_bien'] / x['inscrits'],
            taux_admis_tb = lambda x: x['admis_mention_tres_bien'] / x['inscrits']
        )
    )

    return aggregated[columns_to_show]
```

Dans l’ensemble, on observe une tendance haussière de la médiane, qui affiche sa valeur la plus faible au début de la période analysée (environ 0,8 en 2006) et progresse continuellement par la suite.

Par exemple, la réforme de 2012 a conduit à une légère réduction du taux d'admission dans les deux secteurs, avec une baisse de 1% dans le privé et une diminution inférieure à 1% dans le public.

En revanche, la réforme du collège de 2016 a eu un effet notable sur le taux d'admission de 2017 dans les collèges publics. Pour la première fois depuis 2006, ce taux a dépassé la barre des 0,85. Pendant ce temps, le taux dans le secteur privé est resté relativement stable.

Enfin, le remplacement des épreuves du brevet par le contrôle continu en 2020, en raison de la pandémie de COVID-19, a également eu un impact plus marqué sur le taux d'admission dans les établissements publics que dans les collèges privés.

### Analyse régionale 

```python
data_regions = aggregate_data(data, ['libelle_region'], ['taux_admis']).reset_index()
data_regions.sort_values(by='taux_admis', ascending=False)
```

```python
# les données géographiques des régions françaises
regions = gpd.read_file(data_dir/'a-reg2021.json')

# L'intégration des données géographiques avec les taux d'admission.
regions_taux = regions.merge(data_regions, how='left', left_on='libgeo', right_on='libelle_region')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 5))

# Carte choroplèthe
regions_taux.plot(column='taux_admis', cmap='RdYlBu', linewidth=0.8, edgecolor='black', legend=True, ax=ax)

# Titre de la carte
ax.set_title("Taux d'admission au Brevet par région en France (2006-2021)", fontsize=12, loc='left', pad=15)

# Suppression des axes
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

plt.show()
```

L’analyse révèle que les taux de réussite au Brevet sont les plus faibles dans les départements et régions d’outre-mer. Dans ces territoires, le taux ne dépasse pas 80 %, ce qui les différencie nettement des régions métropolitaines.

La Bretagne se distingue avec un taux de réussite de 90 %, le plus élevé de France. Tandis que la Provence-Alpes-Côte d’Azur affiche le taux le plus bas parmi les régions métropolitaines (83 %).

Compte tenu l’écart significatif constaté entre les taux de réussite dans l’enseignement public et privé lors de l’analyse des données annuelles, il est intéressant d’étudier ces différences à l’échelle régionale pour mieux comprendre leur répartition géographique.

```python
data_regions_secteur = aggregate_data(data, ["libelle_region", "secteur_d_enseignement"], ["taux_admis"])
data_regions_secteur = data_regions_secteur.unstack("secteur_d_enseignement").reset_index()
data_regions_secteur['delta_taux_admis'] = (
    data_regions_secteur[('taux_admis', 'PRIVE')] - data_regions_secteur[('taux_admis', 'PUBLIC')]
)
# Trier les régions en fonction de l’écart entre les taux de réussite dans public et privé
data_regions_secteur.sort_values(by='delta_taux_admis')
```

```python
data_regions_secteur = data_regions_secteur.sort_values(by='delta_taux_admis', ascending=False)

# Définition des positions sur l'axe Y
y_positions = range(len(data_regions_secteur))

fig, ax = plt.subplots(figsize=(10, 7))

# points pour PRIVE
ax.scatter(data_regions_secteur[('taux_admis', 'PRIVE')], y_positions, label="Privé", s=80, edgecolors="black")

# points pour PUBLIC
ax.scatter(data_regions_secteur[('taux_admis', 'PUBLIC')], y_positions, label="Public", s=80, edgecolors="black")

# Lignes reliant PUBLIC à PRIVE avec annotation des différences
for i, region in enumerate(data_regions_secteur.itertuples(index=False)):
    public = getattr(region, "_2")  # PUBLIC
    prive = getattr(region, "_1")   # PRIVE
    delta = getattr(region, "_3")   # Différence
    
    # Tracé de la ligne
    ax.plot([public, prive], [i, i], color="gray", linewidth=1.5, alpha=0.7)
    
    # Ajout de l'annotation au centre de la ligne
    mid_x = (public + prive) / 2
    ax.text(mid_x, i, f"{delta:.2f}", fontsize=10, va='bottom', ha='center', color="#555555")


# Personnalisation des axes
ax.tick_params(colors='#555555')
ax.set_yticks(y_positions)
ax.set_yticklabels(data_regions_secteur['libelle_region'], ha='left') 
ax.yaxis.set_label_position("right") 
ax.yaxis.tick_right()
ax.set_xlim(0.7, 1)

# labels et titre
ax.set_yticklabels(data_regions_secteur['libelle_region'])
ax.set_title("Différence des taux d’admission au Brevet entre le public et le privé (2006 - 2021)", 
             fontsize=16, loc='left', pad=33 )

# légende
ax.legend(loc=2, frameon=False, bbox_to_anchor=(-.028, 1.09), labelcolor='#666666', ncol=2, fontsize=11)

# Suppression de la bordure du graphique
sns.despine(left=True)

# plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.show()
```

Les écarts les plus marqués en termes de qualité de l’éducation sont observés dans les régions d’outre-mer, où la différence entre les taux d’admission des collèges privés et publics varie de 17 % à 23 %.

La Réunion affiche le taux d’admission le plus élevé parmi toutes les régions françaises dans les collèges privés (98 %), bien que l’écart avec le secteur public y soit l’un des plus importants — 20 %.

## Conclusions et recommandations.


Dans le cadre de ce projet, on a analysé les résultats au Diplôme National du Brevet (DNB) entre 2006 et 2021, dans le but d’accompagner les familles de la commune d’Étaules (Charente-Maritime) dans leur choix d’établissement scolaire. L’étude repose sur des données ouvertes mises à disposition par le ministère de l’Éducation nationale via la plateforme data.education.gouv.fr. Ces données, couvrant l’ensemble du territoire national, ne présentaient ni valeurs manquantes ni anomalies. La préparation initiale s’est limitée à l’exclusion des établissements non concernés par l’étude (comme les lycées ou CFA), afin de se concentrer uniquement sur les collèges.

**Analyse nationale**

L’observation des tendances sur quinze années révèle une amélioration continue du taux de réussite au brevet. En 2006, environ 80 % des élèves inscrits obtenaient leur diplôme.  Ce pourcentage a progressé pour atteindre 88 % en 2021. 

Cette tendance a toutefois été ponctuée par des événements spécifiques. La réforme de 2012 a diminué le taux de réussite, tandis que celle de 2016 l'a amélioré, surtout dans le secteur public. En 2020, le recours au contrôle continu dû à la COVID-19 a  provoqué une hausse significative du taux de réussite.

Il convient également de noter des disparités territoriales significatives.  Les départements et régions d’outre-mer affichent généralement des taux de réussite inférieurs à 80 %, alors que certaines régions métropolitaines, comme la Bretagne, atteignent les 90 %. 

L’écart de performance entre les secteurs public et privé reste stable : les établissements privés enregistrent systématiquement de meilleurs résultats que les établissements publics, avec des taux de réussite supérieurs à 90 %. Cet écart est particulièrement prononcé dans les DROM et certaines régions métropolitaines comme l’Île-de-France, les Hauts-de-France ou la région PACA, tandis qu’il est plus réduit en Bretagne ou en Nouvelle-Aquitaine.

**Analyse locale : secteur d’Étaules (17)**

À l’échelle locale, les établissements situés à proximité d’Étaules présentent des profils variés, tant en termes d’effectifs que de résultats. Le collège privé Sainte-Marie Saint-Jean-Baptiste se distinguait historiquement par des taux de réussite supérieurs à 90 %, confirmant la tendance nationale du secteur privé. Toutefois, à partir de 2018, cet établissement connaît un recul relatif, affichant des résultats inférieurs à la moyenne des collèges privés de Nouvelle-Aquitaine.

Dans le même temps, les établissements publics voisins, notamment le collège Fernand Garandeau (La Tremblade) et le collège Henri Dunant (Royan), enregistrent une progression remarquable. Le cas de Fernand Garandeau est particulièrement significatif : son taux de réussite est passé de 70 % en 2006 à 95 % en 2021. Depuis 2019, cet établissement dépasse non seulement le collège privé Sainte-Marie Saint-Jean-Baptiste, mais également les moyennes du secteur public régional et de la région la plus performante, la Bretagne.

En ce qui concerne les mentions obtenues au DNB, si le collège Sainte-Marie Saint-Jean-Baptiste conserve une avance sur les mentions « Bien » et « Très Bien », le collège Fernand Garandeau affiche une forte progression du nombre d'élèves obtenant la mention « Assez Bien » ou étant admis sans mention, ce qui traduit une amélioration globale du niveau des élèves.

**Conclusion**

Les résultats de cette étude montrent que le choix d’un établissement scolaire ne peut reposer uniquement sur son statut public ou privé. À l’échelle nationale comme locale,  les dynamiques propres à chaque collège peuvent influencer considérablement ses performances. 

Dans le secteur d’Étaules, le collège public Fernand Garandeau apparaît aujourd’hui comme un établissement performant et fiable, conjuguant amélioration continue et taux de réussite élevés. 

Pour les familles à la recherche d'un collège de qualité, les données les plus récentes incitent donc à privilégier une évaluation attentive et actualisée de chaque établissement plutôt que de se fier aux généralités ou aux réputations passées.
